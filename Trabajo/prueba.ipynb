```python
lat = ds.variables['latitude'][:]
lon = ds.variables['longitude'][:]
salt = ds.variables['SALT'][:]

print(salt)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import cartopy.crs as ccrs
```

```python
fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection=ccrs.Robinson())
ax.set_global()
plt.contourf(lon, lat, salt[0][0], levels=50, transform=ccrs.PlateCarree())

# ax.coastlines()
ax.gridlines()

plt.colorbar(ax=ax,orientation='horizontal',label='Salinidad (partes por mil)')

plt.show()
```
